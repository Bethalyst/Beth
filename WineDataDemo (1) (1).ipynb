## Visualization

```python
plt.figure(figsize = (10,7))
sns.countplot(x='quality', data = df)
plt.xlabel('Quality of Red Wine')
plt.ylabel('Count')
plt.show()
```

```python
index = 0
labels = df['quality']
features = df.drop('quality', axis=1)

for col in features.items():
    plt.figure(figsize=(10,5))
    sns.barplot(x=labels, y=col[index], data = df, color = "deeppink")
plt.tight_layout()
plt.show()
```

We were looking for any obvious correlations with the column data and the quality

It appears that the lower the acidity and the higher the citric acid, the better the quality of the wine.

```python
fig, ax = plt.subplots(ncols = 6, nrows = 2, figsize=(15,10))
index = 0
ax = ax.flatten()
for col, value in df.items():
    sns.boxplot(y=col, data = df, ax=ax[index])
    index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
plt.show()
```

**Create a distribution plot for all of the categories to view any anomolies or skewness in the distributions**

```python
fig, ax = plt.subplots(ncols = 6, nrows = 2, figsize=(15,10))
index = 0
ax = ax.flatten()
for col, value in df.items():
    sns.distplot(value, ax=ax[index], hist=False)
    index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
plt.show()
```

Based on the distribution and the number of outliers we elected to drop the sulpher dioxide from our data set. There was no real reason to do this other than to show that we can easily drop a column.

## Outlier Removal

Use something like a z-score to remove outliers. We can eliminate the outliers, replace them with the min/max, eliminate the whole row, or set to 'none'.

## Create transforms and filters to change the numeric data to classification data

```python
def densityTransform(density,mean):
    if density > mean:
        return "High"
    else:
        return "Low"
```

```python
mean = df["density"].mean()
```

```python
mean
```

```python
df["density"] = df.apply(lambda row: densityTransform(row["density"],mean), axis = 1)
```
